# Task#4: Data Cleaning/ Feature Engineering

```python
#drop the column 'date
df_alexa = df_alexa.drop(['date', 'rating'],axis=1)
```

```python
#print first and last 5 of dataset now without the dropped column
df_alexa
```

```python
#turn table into binary system
variation_dummies = pd.get_dummies(df_alexa['variation'], drop_first = True)
```

```python
#print binary system
variation_dummies
```

```python
#drop variation column
df_alexa.drop(['variation'], axis=1, inplace=True)
```

```python
#add the encoded column again
df_alexa = pd.concat([df_alexa, variation_dummies], axis=1)
```

```python
#print new encoded set
df_alexa
```

```python
#import libraries
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(df_alexa['verified_reviews'])
```

```python
#define shape of dataset
alexa_countvectorizer.shape
```

```python
#print the type 
type(alexa_countvectorizer)
```

```python
#print
print(vectorizer.get_feature_names())
```

```python
#print
print(alexa_countvectorizer.toarray())  
```

```python
#drop column
df_alexa.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
#concatenate them together
df_alexa = pd.concat([df_alexa, reviews], axis=1)
```

```python
#print
df_alexa
```

```python
#drop the target label coloumns
X = df_alexa.drop(['feedback'],axis=1)
```

```python
# X = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
X
```

```python
y = df_alexa['feedback']
```

# Task #5: Model Building and Model Training

```python
#import libraries for train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=5)
```

```python
#print shape of X train
X_train.shape
```

```python
#print shape of X test
X_test.shape
```

```python
#print shape of y train
y_train.shape
```

```python
#print shape of y test
y_test.shape
```

```python
#import libraries
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

randomforest_classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
randomforest_classifier.fit(X_train, y_train)
```

# Task #6: Model Evaluation

```python
#define ypredict and confusion matrix for train
y_predict_train = randomforest_classifier.predict(X_train)
cm = confusion_matrix(y_train, y_predict_train)
print(cm)
```

```python
#print heatmap for ytrain
sns.heatmap(cm, annot=True)
```

```python
#print classification report for ytrain
print(classification_report(y_train, y_predict_train))
```

```python
#define ypredict and confusion matrix for test
y_predict = randomforest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
print(cm)
```

```python
#print heatmap for test
sns.heatmap(cm, annot=True)
```

```python
#print classification report for test
print(classification_report(y_test, y_predict))
```

# Conclusion

Using the dataset we cleaned the data and created graphs in order to try and understand the patterns and trends within the data in order to draw conclusions. We dropped columns that were not needed as well as created a binary system to help us clearly see which item was being reviewed. 1 showing that this was the item under review and 0 showing that these were not being reviewed. In graph 4 we decided to create a bar plot in order to see the relationship between the variation of Amazon's Alex and the rating that the product receieved. We can see from the graph itself that there is not much of a relationship between any of the variations and how they were rated. For the most part, almost all the variations were given a similar rating. Walnut Finish as well as Oak Finish did have the most number of higher ratings however there are not significantly more higher ratings for Walnut and Oak finish compared to the other variations. 

Once we cleaned our data up. We created a test/train split. We created three different confusion matrixes. The first one was using our predictive data. Looking at this confusion matrix our accuracy, precision and recall were all around 0.99. Having a high precision is important but it does not neccesarily mean that we are getting the correct answer. However, seeing that our accuracy is high, that means we are getting the right answer. This makes sense because we are looking at the data that the model has already analyzed. For the second matrix we can see that the accuracy is around 0.91, recall around 0.7 and precision around 0.92. We can see that compared to the first matrix they are not as precise. This would be because these matrices consist of data that the model has not seen yet. This data is from the test population which the model has not interacted with yet. This error could be due to overfitting which happens when a function corresponds too closely or exactly to a particular set of data. 

For the third and final matrix we can see that the criterion is changed from gini to entropy. Gini and entropy are slightly different. In addition, for this matrix the input has changed to 'length of review', so we can see we're not really on the same exact terms as the last two matrices
